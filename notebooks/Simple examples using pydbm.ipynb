```python
import pydbm
import numpy as np
import seaborn as sns
```

```python
sns.set_context('talk')
sns.set_style('whitegrid')
```

Generate some random binary data sequences with 50% succes



```python
n = 50
data = np.random.rand(n) < 0.5

plt.plot(data)
```

now we make a plot with a BDM with $\alpha = 0.9$ and the prior $\beta(1, 1)$

```python
bdbm = pydbm.BayesianDBM(data, alpha=0.9, prior=(1., 1.))
bdbm.plot_full_posterior_per_trial(interpolation='nearest')
plt.plot(bdbm.get_p_per_trial(), c='white')
```

The parameter that is usually used in model-based neuroimaging studies is p and can be accessed as follows:

```python
bdbm.get_p_per_trial()
```

Now let's try some different parameter settings

```python
bdbm1 = pydbm.BayesianDBM(data, alpha=0.25, prior=(1., 1.))
bdbm2 = pydbm.BayesianDBM(data, alpha=0.25, prior=(1., 5.))
bdbm3 = pydbm.BayesianDBM(data, alpha=0.75, prior=(1., 1.))
bdbm4 = pydbm.BayesianDBM(data, alpha=0.75, prior=(1., 5.))

plt.subplot(2,2,1)
plt.title('alpha 0.25, prior (1, 1)')
bdbm1.plot_full_posterior_per_trial()
plt.plot(bdbm1.get_p_per_trial(), c='white')

plt.subplot(2,2,2)
plt.title('alpha 0.25, prior (1, 5)')
bdbm2.plot_full_posterior_per_trial()
plt.plot(bdbm2.get_p_per_trial(), c='white')

plt.subplot(2,2,3)
plt.title('alpha 0.25, prior (1, 1)')
bdbm3.plot_full_posterior_per_trial()
plt.plot(bdbm3.get_p_per_trial(), c='white')

plt.subplot(2,2,4)
plt.title('alpha 0.25, prior (1, 5)')
bdbm4.plot_full_posterior_per_trial()
plt.plot(bdbm2.get_p_per_trial(), c='white')
```
